```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
df=pd.read_csv("HK Test.csv",delimiter=";")
df.info()
```

```python
#Gregorian Conversion to datetime format
from datetime import date
df['Posting Date'] = pd.to_datetime(df['Posting Date'])
df
```

```python
df['week_number_of_year'] = df['Posting Date'].dt.week
df['month_number_of_year'] = df['Posting Date'].dt.month
df['year'] = df['Posting Date'].dt.year
df
```

```python
df['hijri_calendar']=df['Posting Date']
df
```

```python
from hijri_converter import convert
Hijri=[]
for x in df['hijri_calendar']:
    x=convert.Gregorian(x.year, x.month, x.day).to_hijri()
    Hijri.append(x)
```

```python
for n in Hijri:
    for i in range(0,len(Hijri)):
        n=Hijri[i].dmyformat()
```

```python
#Convert list to np.array
Hijri = np.array(Hijri)
print(type(Hijri))
```

```python
#Convert list to np.array
Hijri = np.array(Hijri)
print(type(Hijri))
```

```python
Islamic_month_number=[]
for i in range(0,len(Hijri)):
    n=Hijri[i].month_name()
    Islamic_month_number.append(n)

print(len(Islamic_month_number))
```

```python
#Convert list to np.array
Islamic_month_number = np.array(Islamic_month_number)
print(type(Islamic_month_number))
```

```python
df['Islamic_month_number']=Islamic_month_number
print(df.head())
df.info()
```

```python
df.drop(['hijri_calendar'], axis='columns', inplace=True)
cleanup_nums={"Islamic_month_number":{"Jumada al-Ula":4,"Jumada al-Akhirah":5,"Rabi’ al-Thani":3,"Sha’ban":7,"Ramadhan":8,"Shawwal":9,"Dhul-Qi’dah":10,"Dhul-Hijjah":11,"Safar":1,"Rajab":6,"Rabi’ al-Awwal":2,"Muharram":0}}
df.replace(cleanup_nums, inplace= True)
df
```

```python
df.info()
```

```python
df=df.groupby(['Posting Date', 'Material Code','week_number_of_year','month_number_of_year','year','Islamic_month_number'])['Act/hl'].sum().reset_index()
df
```

```python
from workalendar.africa import Algeria
cal = Algeria()
years = [2017, 2018, 2019, 2020]
```

```python
dates = df['Posting Date']
DF = pd.DataFrame({'movement_date': dates})
```

```python
DF
```

```python
holidays = pd.DataFrame()
for year in years:
    holidays = holidays.append(pd.DataFrame(cal.get_calendar_holidays(year)))
holidays.columns = ['movement_date', 'holiday']
holidays['movement_date'] = pd.to_datetime(holidays['movement_date'])
DF = DF.merge(holidays, on='movement_date', how='left')
```

```python
DF
```

```python
#DF['holiday_binary'] = DF['holiday'].astype
```

```python
newDF=pd.get_dummies(DF, prefix='holiday', prefix_sep='_', dummy_na=False, columns=['holiday'])
newDF
```

```python
newDF.info()
```

```python
df['holiday_Anniversary of the revolution'] = newDF['holiday_Anniversary of the revolution']
df['holiday_Ashura'] = newDF['holiday_Ashura']
df['holiday_Eid al-Adha'] = newDF['holiday_Eid al-Adha']
df['holiday_Independence Day'] = newDF['holiday_Independence Day']
df['holiday_Islamic New Year'] = newDF['holiday_Islamic New Year']
df["holiday_Prophet's Birthday"] = newDF["holiday_Prophet's Birthday"]
df.info()
```

```python
df['holiday_Ashura'] = df['holiday_Ashura'].astype('int')
df['holiday_Anniversary of the revolution'] = df['holiday_Anniversary of the revolution'].astype('int')
df['holiday_Eid al-Adha']=df['holiday_Eid al-Adha'].astype('int')
df['holiday_Independence Day']=df['holiday_Eid al-Adha'].astype('int')
df['holiday_Islamic New Year']=df['holiday_Eid al-Adha'].astype('int')
df["holiday_Prophet's Birthday"]=df['holiday_Eid al-Adha'].astype('int')
df.info()
```

```python
df.drop(['Posting Date'], axis=1)
```

```python
#Correlation
pd.DataFrame(df.corr()['Act/hl']).sort_values(by='Act/hl',ascending=False)
```

```python
#extract x and y from our data
x=df[['week_number_of_year','month_number_of_year',"week_number_of_year","Islamic_month_number",'holiday_Anniversary of the revolution','holiday_Ashura','holiday_Eid al-Adha','holiday_Independence Day','holiday_Islamic New Year',"holiday_Prophet's Birthday"]]  #we have more than one input
y=df["Act/hl"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=40) #splitting data with test size of 30%

model=LinearRegression() #build linear regression model
model.fit(x_train,y_train) #fitting the training data
predicted=model.predict(x_test) #testing our model’s performance

print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
a=model.coef_
b=model.intercept_
print(a)
print(b)
```
